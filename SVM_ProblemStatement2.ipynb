month- month of the year: 'jan' to 'dec'

rain- outside rain in mm/m2 : 0.0 to 6.4
Size_Categorie- the burned area of the forest ( Small , Large)

```python
import pandas as pd 
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

### Data Preprocessing

```python
#Dropping columns which are not required
forest_data=forest_data.drop(columns=['dayfri','daymon','daysat','daysun','daythu','daytue','daywed','monthapr',	
                               'monthaug','monthdec','monthfeb','monthjan','monthjul','monthjun','monthmar',
                               'monthmay','monthnov','monthoct','monthsep'],axis=1)
```

```python
forest_data
```

```python
forest_data["month"].value_counts()
```

```python
month_data={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
forest_data.replace(month_data,inplace=True)
```

```python
forest_data
```

```python
forest_data["day"].value_counts()
```

```python
day_data={'day':{'mon':1,'tue':2,'wed':3,'thu':4,'fri':5,'sat':6,'sun':7}}
forest_data.replace(day_data,inplace=True)
```

```python
forest_data
```

```python
forest_data['size_category'].unique()
```

```python
forest_data.size_category.replace(('small','large'),(1,0),inplace=True)
```

```python
forest_data
```

```python
forest_data.dtypes
```

```python
X=forest_data.iloc[:,0:11]
X
```

```python
Y=forest_data["size_category"]
Y
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(X)
x_scaled
```

```python
scaled_forest_data=pd.DataFrame(x_scaled,columns=X.columns)
scaled_forest_data
```

### Model Building

```python
X_transformed=scaled_forest_data
X_transformed
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_transformed,Y,test_size=0.20,random_state=12,stratify=Y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

### Model Training

```python
svc_model=SVC()
```

```python
svc_model.fit(X_train,Y_train)
```

### Model Testing

```python
y_train_pred=svc_model.predict(X_train)
y_train_pred
```

```python
y_test_pred=svc_model.predict(X_test)
y_test_pred
```

### Model Evaluation

Training Data

```python
print(confusion_matrix(Y_train,y_train_pred))
```

```python
print(classification_report(Y_train,y_train_pred))
```

Testing Data

```python
print(confusion_matrix(Y_test,y_test_pred))
```

```python
print(classification_report(Y_test,y_test_pred))
```
